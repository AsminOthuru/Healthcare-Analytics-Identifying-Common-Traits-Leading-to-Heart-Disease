# Handling Missing Values

# Sanity Checks

```python
# performing sanity checks.
def sanity_checks(df):
    # Define the valid ranges for each column
    valid_ranges = {
        "HeartDiseaseorAttack": (0, 1),
        "HighBP": (0, 1),
        "HighChol": (0, 1),
        "CholCheck": (0, 1),
        "BMI": (18, 40),
        "Smoker": (0, 1),
        "Stroke": (0, 1),
        "Diabetes": (0, 2),
        "PhysActivity": (0, 1),
        "Fruits": (0, 1),
        "Veggies": (0, 1),
        "HvyAlcoholConsump": (0, 1),
        "AnyHealthcare": (0, 1),
        "NoDocbcCost": (0, 1),
        "GenHlth": (1, 5),
        "MentHlth": (0, 30),
        "PhysHlth": (0, 30),
        "DiffWalk": (0, 1),
        "Sex": (0, 1),
        "Age": (1, 13),
        "Education": (1, 6),
        "Income": (1, 6)
    }

    # Check each column against its valid range
    for column, (min_val, max_val) in valid_ranges.items():
        if not df[column].between(min_val, max_val, inclusive='both').all():
            print(f"No: {column} column contains out-of-range values")
            return

    print("sanity")

sanity_checks(df_new)
```

##### From above we found that there are outliers in BMI column lets handle them.

# Handling Outliers

```python
# Calculate quartiles and IQR.
Q1 = df_new['BMI'].quantile(0.25)
Q3 = df_new['BMI'].quantile(0.95)
IQR = Q3 - Q1

# Define outliers based on IQR.
outliers = df_new[(df_new['BMI'] < Q1 - 1.5 * IQR) | (df_new['BMI'] > Q3 + 1.5 * IQR)]

# Check if outliers are present and print a message.
if outliers.empty:
    print("No outliers found.")
else:
    print("Outliers are present.")
```

```python
# Box plot for outlier detection.
plt.figure(figsize=(8, 6))
df_new.boxplot(column=['BMI'])
plt.show()
```

```python
# Removing Outliers.
df_cleaned = df_new[(df_new['BMI'] >= Q1) & (df_new['BMI'] <= Q3)]
```

# BiVariate Analysis

```python
# Scatter Plots for Numerical Variables.

plt.figure(figsize=(12, 6))

plt.subplot(2, 3, 1)
sns.scatterplot(x='BMI', y='GenHlth', data=df_cleaned)
plt.title('BMI vs General Health')

plt.subplot(2, 3, 2)
sns.scatterplot(x='Age', y='MentHlth', data=df_cleaned)
plt.title('Age vs Mental Health')

plt.subplot(2, 3, 3)
sns.scatterplot(x='Income', y='PhysHlth', data=df_cleaned)
plt.title('Income vs Physical Health')

plt.tight_layout()
plt.show()
```

The three scatter plots compare specific health metrics against each other: "BMI vs General Health," "Age vs Mental Health," and "Income vs Physical Health." Each plot shows the spread of data points, with potential trends or clusters indicating relationships. For instance, the uniform spread in "BMI vs General Health" suggests a lack of strong correlation, while other plots may hint at more complex relationships. These visualizations are key in identifying factors that influence health outcomes, guiding more detailed statistical analyses.

```python
# Define target and other columns.
target_column = "HeartDiseaseorAttack"
other_columns = ["HighChol", "Income", "Education", "HighBP", "Age", "Diabetes", "Smoker", "HvyAlcoholConsump"]

# Number of columns and rows for the subplot grid.
n_cols = 3
n_rows = (len(other_columns) + n_cols - 1) // n_cols  # Calculate rows needed.

# Create subplots
fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, 10))
axes = axes.flatten()  # Flatten to easily iterate over them.

# Loop through other columns and plot line plots in subplots.
for i, column in enumerate(other_columns):
    sns.lineplot(x=column, y=target_column, data=df_cleaned, ax=axes[i])
    axes[i].set_title(f"{column} vs. {target_column}")

# Hide any empty subplots if there are any.
for j in range(i+1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

The line graphs show the relationship between specific health metrics (e.g., high cholesterol, high blood pressure, diabetes, smoking) and the likelihood of a heart disease attack. Positive correlations are evident, indicating that as these risk factors increase, so does the incidence of heart disease. Interestingly, higher income and education levels are associated with a lower risk of heart disease, while age shows a clear positive correlation, underscoring the importance of targeting these areas in heart disease prevention efforts.

```python
# Heatmap of Correlations.

plt.figure(figsize=(10, 8))
sns.heatmap(df_cleaned[['BMI', 'GenHlth', 'MentHlth', 'PhysHlth', 'Age', 'Income']].corr(), annot=True, cmap='coolwarm', fmt='.2f', vmin=-1, vmax=1)
plt.title('Correlation Heatmap for Numerical Variables')
plt.show()
```

The correlation heatmap illustrates the strength and direction of linear relationships between the numerical variables, with values ranging from -1 to 1. Strong positive correlations, such as between "GenHlth" and "PhysHlth" (0.50), suggest that as one variable increases, the other does as well. Conversely, negative correlations, like between "Income" and "GenHlth" (-0.32), indicate an inverse relationship. The heatmap is color-coded, with red indicating positive and blue negative correlations, offering a quick overview of potential predictors or multicollinearity issues.

```python
# Define a function to create pie charts
def plot_pie_chart(column, title):
    data = df_cleaned[column].value_counts()
    labels = data.index
    sizes = data.values

    plt.figure(figsize=(3, 3))
    plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
    plt.title(title)
    plt.axis('equal')
    plt.show()

# Pie charts for categorical variables
plot_pie_chart('Smoker', 'Smoking Status')
plot_pie_chart('PhysActivity', 'Physical Activity')
plot_pie_chart('Fruits', 'Fruit Consumption')
plot_pie_chart('Veggies', 'Vegetable Consumption')
```

```python
# Grouped bar charts
categorical_vars = ['Smoker', 'PhysActivity', 'Fruits', 'Veggies']
plt.figure(figsize=(10, 6))

for i, var in enumerate(categorical_vars, 1):
    plt.subplot(2, 2, i)
    sns.countplot(x=var, hue='HeartDiseaseorAttack', data=df_cleaned)
    plt.title(f'Heart Disease by {var}')

plt.tight_layout()
plt.show()
```

The bar charts illustrate the relationship between heart disease and factors like smoking, physical activity, fruit and vegetable consumption, and their respective distributions. Smoking is associated with a higher risk of heart disease, as indicated by a higher incidence among smokers. Lack of physical activity is also a significant risk factor, with more heart disease cases among inactive individuals. Conversely, regular consumption of fruits and vegetables shows a potential protective effect against heart disease. The graphs emphasize the importance of lifestyle modifications in preventing heart disease.
